## <font color=#33A5FF>****Análisis de Datos English Premier League (EPL) 2020-2021****<tfont>

```python
#Librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Cargamos los **datos** para luego ***limpiar nuestros datos***.

```python
#cargamos un archivo CSV separado por comas
epl = pd.read_csv('EPL_20_21.csv')
```

```python
#leer 5 primeras lineas
epl.head()
```

```python
epl.shape
```

```python
len(epl)
```

```python
#leer 4 primeras lineas aleatorias
epl.sample(4)
```

```python
#Información del DataFrame
epl.info()
```

```python
#Resumen de Datos estadisticos ---- Solo columnas numéricas
epl.describe().T
```

```python
#Número total de valores nulos
epl.isna().sum()
```

```python
epl.columns
```

Creamos 2 columnas mas **Minutos por partido** y ***Goles por partido***.

```python
# creacion de 3 columnas
epl["MinsPerMatch"]=(epl["Mins"]/epl["Matches"]).astype(int)
epl["GoalsPerMatch"]=(epl["Goals"]/epl["Matches"]).astype(float)
epl["AssistsPerMatch"]=(epl["Assists"]/epl["Matches"]).astype(float)
```

```python
#visualizamos las nuevas columnas
epl.sample(5)
```

```python
epl.shape
```

```python
epl.columns
```

```python
#Total de goles--temporada EPL 2020-2021 ultima temporada
TotalGoals=epl["Goals"].sum()
TotalAssists=epl["Assists"].sum()
TotalGoals
```

```python
#Total de goles--temporada EPL 2020-2021 ultima temporada
TotalAssists=epl["Assists"].sum()
TotalAssists
```

```python
#Goles x Penales ---- temporada EPL 2020-2021 ultima temporada
TotalPenaltyGoals=epl["Penalty_Goals"].sum()
TotalPenaltyGoals
```

```python
# Total Penales ---- temporada EPL 2020-2021 ultima temporada
TotalPenaltyAttempts=epl["Penalty_Attempted"].sum()
TotalPenaltyAttempts
```

```python
# Total Penales Fallados ---- temporada EPL 2020-2021 ultima temporada
TotalPenaltyFailed=[TotalPenaltyAttempts - TotalPenaltyGoals]
TotalPenaltyFailed
```

### Gráfico Pie chart **Penales Marcados** y **Penales Atajados**.

```python
# Gráficos de penales
plt.figure(figsize=(13,6))
pl_not_Scored=TotalPenaltyAttempts-TotalPenaltyGoals
data=[pl_not_Scored,TotalPenaltyGoals]
nom=["Penales atajados","Penales marcados"]
colores=sns.color_palette("Paired")
plt.pie(data,labels=nom,colors=colores,autopct="%.0f%%")
plt.show()
```

```python
# Posiciones únicas de cada jugador 
# Portero (GK), defensor (DF), centrocampista (MF) delantero (FW)
epl["Position"].unique()
```

```python
# Filtrar jugadores delanteros (FW)
epl[epl["Position"]=="FW"]
```

```python
# Filtrar jugadores delanteros y defensas (FW,DF)
epl[epl["Position"]=="FW,DF"]
```

```python
# Cantidad jugadores delanteros (FW)
epl_fw=epl[epl["Position"]=="FW"]
epl_fw["Name"].count()
```

```python
#  Nacionalidad de los Jugadores
np.size((epl["Nationality"].unique()))
```

```python
# Nacionalidad de la mayoría de los jugadores
nationality=epl.groupby("Nationality").size().sort_values(ascending=False)
nationality.head(10).plot(kind="bar",figsize=(12,6),color=sns.color_palette("mako"))
```

```python
type(nationality)
```

```python
# Los Club con maximo numero de jugadores en la temporada
epl["Club"].value_counts().nlargest(5).plot(kind="bar",color=sns.color_palette("hls",8))
plt.show()
```

```python
# Los Club con menor numero de jugadores en la temporada
epl["Club"].value_counts().nsmallest(5).plot(kind="bar",color=sns.color_palette("rocket_r"))
plt.show()
```

### Edad de los jugadores

```python
# Grupo de edades de los Jugadores
Under20=epl[epl["Age"]<=20]
age20_25=epl[(epl["Age"]>20)&(epl["Age"]<=25)]
age25_30=epl[(epl["Age"]>25)&(epl["Age"]<=30)]
Above30=epl[epl["Age"]>30]
```

```python
x=np.array([Under20["Name"].count(),age20_25["Name"].count(),
          age25_30["Name"].count(),Above30["Name"].count()])
mylabels=["<=20",">20 & <=25",">25 & <=30",">30"]
plt.figure(figsize=(13,6))
plt.title("Total de jugadores por grupos de edad",fontsize=18)
plt.pie(x,labels=mylabels,autopct="%.1f%%")
plt.show()
```

```python
# Total de jugadores menores de 20 años en cada club
PlayerUnder20=epl[epl["Age"]<20]
PlayerUnder20["Club"].value_counts().plot(kind="bar",color=sns.color_palette("cubehelix"))
plt.show()
```

```python
# Jugadores menores de 20 años en  club Manchester United 
PlayerUnder20[PlayerUnder20["Club"]=="Manchester United"]
```

```python
# Jugadores menores de 20 años en  club Chelsea
PlayerUnder20[PlayerUnder20["Club"]=="Chelsea"]
```

```python
# Gráfico BoxPlot de la edad promedio de los jugadores de cada club
plt.figure(figsize=(10,10))
sns.boxplot(x="Club",y="Age",data=epl)
plt.xticks(rotation=60)
plt.show()
```

```python
# Clubs con Edades promedio mas altas
NumPlayer=epl.groupby("Club").size()
data=(epl.groupby("Club")["Age"].sum())/NumPlayer
data.sort_values(ascending=False)
```

```python
# Grafica de Total de asistencias gol por cada club
AssitsByClub=pd.DataFrame(epl.groupby("Club",as_index=False)["Assists"].sum())
sns.set(style="whitegrid",color_codes=True)
ax=sns.barplot(x="Club",y="Assists",data=AssitsByClub.sort_values(by="Assists"))
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Asistencias",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,10)
plt.title("Grafica de Club vs total de asistencias de gol",fontsize=20)
plt.show()
```

```python
# Jugadores top de 10 Asistencias Gol
top_10_assists=epl[["Name","Club","Assists","Matches"]].nlargest(n=10,columns="Assists")
top_10_assists
```

```python
# Jugadores top de 10 con mas pases intentados
top_10_PassesA=epl[["Name","Club","Passes_Attempted","Matches"]].nlargest(n=10,columns="Passes_Attempted")
top_10_PassesA
```

```python
# Jugadores top 10 con menos pases intentados
top_10_NPasses=epl[["Name","Club","Passes_Attempted","Matches"]].nsmallest(n=10,columns="Passes_Attempted")
top_10_NPasses
```

```python
# Grafica del total de goles de cada club
GoalsByClubs=pd.DataFrame(epl.groupby("Club",as_index=False)["Goals"].sum())
sns.set(style="whitegrid",color_codes=True)
ax=sns.barplot(x="Club",y="Goals",data=GoalsByClubs.sort_values(by="Goals"))
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goles",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,10)
plt.title("Grafica de Club vs total de goles ",fontsize=20)
plt.show()
```

```python
#Top 10 de  Goles por Jugador
top_10_goals=epl[["Name","Club","Goals","Matches"]].nlargest(n=10,columns="Goals")
top_10_goals
```

```python
# Top 10 Goles por partido
top_10_goals_per_match=epl[["Name","GoalsPerMatch","Matches",
                            "Goals"]].nlargest(n=10,columns="GoalsPerMatch")
top_10_goals_per_match
```

```python
epl.columns
```

```python
# Grafica Pi Chart - Goles con asistencia y Goles sin asistencia
plt.figure(figsize=(20,10))
assists=epl["Assists"].sum()
data=[TotalGoals-assists,assists]
nom=["Goles sin Asistencia","Goles con asistencias"]
colores=sns.color_palette("Set1")
plt.pie(data,labels=nom,colors=colores,autopct="%.0f%%")
plt.show()
```

```python
#Top 10 jugadores con mas tarjetas amarillas
epl_yellow=epl.sort_values(by="Yellow_Cards",ascending=False)[:10]
plt.figure(figsize=(20,10))
plt.title("Top 10 jugadores con mas tarjetas amarillas", fontsize=20)
c=sns.barplot(x=epl_yellow["Name"],y=epl_yellow["Yellow_Cards"],
             label="Players",color="yellow")
plt.ylabel("Numero de tarjetas amarillas")
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
plt.show()
```

```python
#Top 10 jugadores con mas tarjetas rojas
#Top 10 jugadores con mas tarjetas amarillas
epl_red=epl.sort_values(by="Red_Cards",ascending=False)[:10]
plt.figure(figsize=(20,10))
plt.title("Top 10 jugadores con mas tarjetas rojas", fontsize=20)
c=sns.barplot(x=epl_red["Name"],y=epl_red["Red_Cards"],
             label="Players",color="red")
plt.ylabel("Numero de tarjetas Rojas")
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
plt.show()
```
